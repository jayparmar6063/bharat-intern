```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('movie.csv')
movies.head()
```

```python
movies.columns
```

```python
movies.info()
```

```python
movies['tag'] = movies['genre'] + movies['overview']
```

```python
movies.head()
```

```python
movies = movies[['id','title','tag']]
```

```python

movies.head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vec = cv.fit_transform(movies['tag'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
sim
```

```python
name = 'Your name.'
name = name.lower()
movies[movies['title'].str.lower()  == name ]
```

```python
dist = sorted(list(enumerate(sim[9])),reverse=True, key= lambda vec:vec[1])
```

```python
dist
```

```python
for i in dist[0:5]:
    print(movies.iloc[i[0]].title)
```

```python
def recommend(movie_name):
    movie_name = movie_name.lower()
    index = movies[movies['title'].str.lower() == movie_name].index[0]
    distance = sorted(list(enumerate(sim[index])),reverse=True, key= lambda vec:vec[1])
    for i in distance[0:5]:
        print(movies.iloc[i[0]].title)
```

```python
recommend("your name.")
```
